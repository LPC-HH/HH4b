```python
import mplhep as hep
import matplotlib.pyplot as plt
import os

plt.style.use(hep.style.CMS)
```

```python
# make two empty canvases side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# label axes
ax1.set_xlabel(r"$D$(ggHH-vs-bkg)")
ax2.set_xlabel(r"$D$(qqHH-vs-bkg)")
ax1.set_ylabel(r"$T_{Xbb}(H_2)$")
ax2.set_ylabel(r"$T_{Xbb}(H_2)$")

# fill in polygons
ax1.fill_between([0.03, 1], [0.3, 0.3], y2=[0.85, 0.85], color="#9c9ca1", alpha=1)
ax1.fill_between([0.03, 0.94], [0.85, 0.85], y2=[0.945, 0.945], color="#f89c20", alpha=1)
ax1.fill_between([0.03, 0.755], [0.945, 0.945], y2=[1, 1], color="#f89c20", alpha=1)
ax1.fill_between([0.755, 0.94], [0.945, 0.945], y2=[1, 1], color="#5790fc", alpha=1)
ax1.fill_between([0.94, 1], [0.85, 0.85], y2=[0.945, 0.945], color="#5790fc", alpha=1)
ax1.fill_between([0.94, 1], [0.945, 0.945], y2=[1, 1], color="#e42536", alpha=1)
ax2.fill_between([0.9825, 1], [0.8, 0.8], y2=[1, 1], color="#964a8b", alpha=1)
ax2.fill_between([0.03, 0.9825], [0.8, 0.8], y2=[1, 1], color="#9c9ca1", alpha=1)
ax2.fill_between([0.03, 1], [0.3, 0.3], y2=[0.8, 0.8], color="#9c9ca1", alpha=1)

# add dotted lines
ax1.plot([0.755, 0.755], [0.945, 1], color="k", linestyle="--")
ax1.plot([0.94, 0.94], [0.85, 1], color="k", linestyle="--")
ax1.axhline(y=0.85, color="k", linestyle="--")
ax1.plot([0.755, 1], [0.945, 0.945], color="k", linestyle="--")
ax2.plot([0.9825, 0.9825], [0.8, 1], color="k", linestyle="--")
ax2.plot([0.9825, 1], [0.8, 0.8], color="k", linestyle="--")

# add text labels
ax1.text(0.47, 0.88, "ggHH SR 2", transform=ax1.transAxes, fontsize=20, ha="center")
ax1.text(0.47, 0.65, "ggHH SR 3", transform=ax1.transAxes, fontsize=20, ha="center")
ax1.text(0.89, 0.84, "ggHH SR 1", transform=ax1.transAxes, fontsize=20, ha="center", rotation=45)
ax1.text(0.89, 0.6, "ggHH SR 2", transform=ax1.transAxes, fontsize=20, ha="center", rotation=45)
ax1.text(0.47, 0.25, "QCD CR", transform=ax1.transAxes, fontsize=20, ha="center")
ax2.text(0.97, 0.6, "qqHH SR", transform=ax2.transAxes, fontsize=20, ha="center", rotation=90)
ax2.text(0.47, 0.47, "QCD CR", transform=ax2.transAxes, fontsize=20, ha="center")

# add broken axis lines
d = 0.5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(
    marker=[(-1, -d), (1, d)],
    markersize=12,
    linestyle="none",
    color="k",
    mec="k",
    mew=1,
    clip_on=False,
)

for ax in (ax1, ax2):
    # set x and y limits
    ax.set_xlim(0.7, 1)
    ax.set_ylim(0.7, 1)
    # plot broken line
    ax.plot([0, 1], [0.095, 0.095], transform=ax.transAxes, **kwargs)
    ax.plot([0, 1], [0.105, 0.105], transform=ax.transAxes, **kwargs)
    ax.plot(0.095, 0, transform=ax.transAxes, **kwargs)
    ax.plot(0.105, 0, transform=ax.transAxes, **kwargs)
    # change tick mark labels
    ax.set_yticks([0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0])
    ax.set_yticklabels(["0.3", "0.75", "0.80", "0.85", "0.90", "0.95", "1.0"])
    ax.set_xticks([0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0])
    ax.set_xticklabels(["0.03", "0.75", "0.80", "0.85", "0.90", "0.95", "1.0"])
ax.set_xticklabels(["0.0", "0.75", "0.80", "0.85", "0.90", "0.95", "1.0"])

plt.tight_layout()

os.makedirs("./plots", exist_ok=True)
plt.savefig("./plots/boosted_categories.pdf")
```
